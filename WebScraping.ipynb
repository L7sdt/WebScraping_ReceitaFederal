```
#import used libs and the html file
from urllib.request import urlopen
from bs4 import BeautifulSoup
html = urlopen("https://www.gov.br/receitafederal/pt-br/assuntos/agenda-tributaria/agenda-tributaria-2021/agenda-tributaria-janeiro-2021/dia-06-01-2021")
bs = BeautifulSoup(html, 'html.parser')
```

```
#the html prettified
print(bs.prettify())
```

```
# Search with single class
#linhas = bs.find_all('tr', {'class':'even'})
#Search with multiple classes
#linhas = bs.find_all("tr",  {"class": ["odd", "even"]})

#get tbody tag content
linhas = bs.find_all('tbody')
```

```
## Print text in every line ##
for i in linhas:
    print(i)
```

```
## print text in every tag child ##
filhasTR = []
filhasTD = []
for i in linhas:
    filhasTR = i.findChildren("tr")
    print(filhasTR)
```

```
## get td from every tr ##
## print text in every tag child ##
for i in filhasTR:
    filhasTD = i.findChildren("td")
    print(filhasTD[0])
    print(filhasTD[1])
    print(filhasTD[2])

```

```
#append td content in lists
codigo, descricao, periodo = [], [], []
for i in filhasTR:
    children = i.findChildren("td")
    codigo.append(children[0].text)
    descricao.append(children[1].text)
    periodo.append(children[2].text)
```

```
#using panda to create dataframe from lists above
import pandas as pd
df = pd.DataFrame({'Código': codigo, 'Descrição': descricao, 'Período': periodo})

display(df)
```
